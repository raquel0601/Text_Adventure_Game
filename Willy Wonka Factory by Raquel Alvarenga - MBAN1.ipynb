```python
#importing packages
import emoji #emoji package to make the game more visually appealing.
import random #selects numbers and objects on lists at random in this game.
import colorama #adding some color in the 'Press enter' sections.
from colorama import Fore

######## Game Introduction ######

game_intro= print(f"""

       Charlie and the Chocolate Factory! {emoji.emojize(':chocolate_bar:'*5)}
{'-' * 100}
       This text game is about Charlie and the Chocolate Factory, a 2005 musical fantasy
       film directed by Tim Burton and written by John August, based on the 1964 British 
       novel of the same name by Roald Dahl. The main objective of this game is to entertain
       and guide the user around the chocolate factory to pass different challenges and win
       a prize. There are 4 rooms, and the structure looks as follows with its respective order:

       >> Main_door
       >> Candy_room
       >> OompaLoompa_room
       >> Final_room     
          >Win or Fail
       
       Each room has a different activity and the user cannot fail the game before getting 
       to the Final_room. 
       
{'-' * 100}\n
    
""")

######## Game Starts Here ######

input('\033[1;31m\t<Press enter or return key to start the game>\t')

#Obtaining more information about the user.

user_name = input(prompt= f"""
        First things first! Dear friend, what is your name?     
        >> """).title() #adding for case control

place = input(prompt= f"""
        Great to finally meet you in person {user_name}! 
        Where are you visiting us from?                         
        >> """).title() #adding for case control

### First Stop ###
def main_door():
    """This is the main door to Willy Wonka's Factory""" #docstring
    
    print(f"""
        Thank you so much for coming all the way from {place}! Please, open  
        the door to enter the factory. Mr. Wonka is waiting for you inside.\t\n\n""")
    
    input('\033[1;31m\t<Press enter or return key to open the door>\t')
    
    
    candy_room() 
    
### Second stop ###
def candy_room():
    """This is the Candy Room, where the user can delight with a candy. He or She is
    then assigned a superpower at random based on their choice."""
    
#List with super powers
#Super power idented this way so its aligned with my paragraph in candy_room.
    super_power= ["superhuman speed.",
"""shrinking powers. You‚Äôll be able to become small 
       like my friends Oompa-loompas.""",
"""x-ray vision. You‚Äôll be able to see through objects and 
       detect them in the darkness.""",
"""freezing powers. You‚Äôll be able to freeze objects and 
       other people"""]

#Prompt to choose a candy or sweet item.
    print(f"""
{'-' * 100}
    
        Welcome to my factory {user_name}! I‚Äôm Willy Wonka and I‚Äôm very pleased to 
        have you. I guess you must be very hungry from your trip. Please, treat 
        yourself and choose a candy from the Candy Room to start your journey. 
        I promise, you‚Äôll need the energy! {emoji.emojize(':collision:')}

        Please, choose one of the following candies:
        A. Peanut Butter Taffy {emoji.emojize(':candy:')}
        B. Magic Wonka Chocolate Bar {emoji.emojize(':chocolate_bar:')}
        C. Galactic Ice Cream {emoji.emojize(':ice_cream:')}
        D. Crickets and Sugar cookie {emoji.emojize(':cookie:')}
          
{'-' * 100}\n

      """)
      
    #asking user for candy choice and giving them a superpower at random.  
    choice= input("\033[1;31m\t<<Select a candy by chosing A,B,C, or D >>\t").title()

    #used random.choice in all conditional statements to assign a superpower.
    if choice == "A":
        print(f"""
       Here is your Peanut Butter Taffy {emoji.emojize(':candy:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    
    elif choice == "B":
        print(f"""
       Here is your Magic Wonka Chocolate Bar {emoji.emojize(':chocolate_bar:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    
    elif choice == "C":
        print(f"""
       Here is your Galactic Ice cream {emoji.emojize(':ice_cream:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    
    elif choice == "D":
        print(f"""
       Here is your Crickets and Sugar cookie {emoji.emojize(':cookie:')}! 
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    elif "Peanut" in choice or "Butter" in choice or "Taffy" in choice:
        print(f"""
       Here is your Peanut Butter Taffy {emoji.emojize(':candy:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    elif "Magic" in choice or "Wonka" in choice or "Chocolate" in choice or "Bar" in choice:
        print(f"""
       Here is your Magic Wonka Chocolate Bar {emoji.emojize(':chocolate_bar:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    elif "Galactic" in choice or "Ice" in choice or "Cream" in choice:
        print(f"""
       Here is your Galactic Ice Cream {emoji.emojize(':ice_cream:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    elif "Crickets" in choice or "Sugar" in choice or "Cookie" in choice:
        print(f"""
       Here is your Crickets and Sugar cookie {emoji.emojize(':cookie:')}!
       
       Great selection {user_name}! What you did not know is that each of my candy 
       provides you with a super power for the next couple of hours. Based on your
       selection, you got {random.choice(super_power)}. 
       
       Pretty cool, Huh? {emoji.emojize(':smirk:',use_aliases=True)}
    """)
    else:
        print("""
       Oops! We do not have that candy at the moment, please try again!""")
        candy_room() #starts the room again if the user types something wrong.
        
    input('\n\033[1;31m\t<Press enter or return key to continue>\t\n')
        
    OompaLoompa_room()

### Third Stop ###

#User needs to find the Oompa-Loompa with a key.
def OompaLoompa_room():
    f"""This is the Oompa-Loompa Room, where the user has 3 chances to 'guess' a number
    and find the Oompa-Loompa with a golden key in the pocket. If successful, they get 
    a year's supply of their favorite candy.""" #Docstring
    
    print(f"""
{'-' * 100}

       It‚Äôs now time to have some fun {user_name}! {emoji.emojize(':dizzy:')}

       Get in the boat {emoji.emojize(':sailboat:')}! We‚Äôll go through the Chocolate River and head to the 
       whipped cream room, where I will introduce you to my Oompa-Loompas.

       Oompa-Loompas run this factory and prepare candies for millions of kids worldwide.
       I got them 10 years ago from Loompaland and offered them unlimited cocoa beans
       to come and live with me in the factory {emoji.emojize(':factory:',use_aliases=True)}. They are small but extremely strong and
       humorous. They also love to sing but are a bit shy about it when there are strangers
       around {emoji.emojize(':flushed:',use_aliases=True)}.

       Wow, it seems like we just arrived! There are {emoji.emojize(':one::five:',use_aliases=True)}Oompa-Loompas in this room,
       and one of them has a golden key in the pocket. Find the Oompa-Loompa holding 
       that {emoji.emojize(':key:')}, and you will get a year‚Äôs supply of your favorite candy!
       
       You will have 3 chances to guess the right Oompa-Loompa number. Bring me that key
       and the candy will be yours! {emoji.emojize('Good Luck :blush:!',use_aliases=True)} 
       
{'-' * 100}  

    """)
    #using random.randint to pick the winning number at random.
    number= random.randint(1,15) 
    #the user has 3 guesses before loosing.
    guesses= 3
    
    #While loop and conditional statement 
    while guesses >0 :
        
    # this repeats until the user doesn't have more chances or guess correctly.
        choice = input("""
    \t What is your guess (1-15)? >>\t""")

    # converting choice for numerical comparison
        choice = int(choice)
    
    # conditional statement based on user input
        if choice == number:
            print(f"""
            You did it! You won a year's supply of your favorite candy!{emoji.emojize(':raised_hands::lollipop:',use_aliases=True)}
            """)
            break #it breaks the choice input if the user guesses correctly.

        elif choice != number:
            guesses -= 1
        
            if guesses >0:
                print(f"""
        That's incorrect. You have {guesses} guesses remaining.""")
            
            elif guesses ==0:
                print(f"""
{'-' * 100}
        
        You are out of guesses. The Oompa-Loompa holding the right key was number {number}.
                      
        I know you might be a bit disappointed, but there are more surprises coming up! {emoji.emojize(':wink:',use_aliases=True)}
{'-' * 100}\n
         """)               
        else: 
            print("Error")
        
        input('\033[1;31m\t<Press enter or return key to continue>\t')
    
    
    final_room()
        
### Fourth Stop and last room of this game ###        
def final_room():
    f""" This is the final room in the Willy Wonka's Factory. The user has to guess
    the right answer in order to win the game and not be thrown away in the garbage chute.
    """ #docstring
    print(f""" 
{'>' * 100}
    
      Hello {user_name}, you have done it great so far! You're about to go the last 
      room of this adventure. {emoji.emojize(':cyclone:',use_aliases=True)}
      
      It's sad to let you go, but your loved ones are waiting for you in the real
      world {emoji.emojize(':cry:',use_aliases=True)}. You will now have to spin the Vanilla Wheel of Fortune to reveal the 
      last room we'll be visiting in my factory.
    
{'>' * 100}
    """)
    
#objects for while loop.  
    surprise= "Nut Sorting Room"
    presses = 2
    
    while presses > 0:
        input("""\033[1;31m\t
        <Press enter twice to spin the Wheel of Fortune>\n""")
        presses -= 1
    
    
    if surprise == "Nut Sorting Room":
        final_choice= input(prompt= f"""
{'-' * 100}
   
     Welcome to the Nut Sorting Room!{emoji.emojize(':chestnut:',use_aliases=True)} There are thousands of squirrels working in this room. 
     Their job is to peel the good nuts and get rid of the bad nuts.
 
     The squirrels think you are a bad nut, and they want to throw you on the garbage chute. What
     are you going to do?

     A. Try to escape. 
     B. Negotiate.
     C. Fight back.
     
     <<Select an option by choosing A,B, or C >>
     
{'-' * 100}
     
 \t""").title()
        
        if "A" in final_choice or "Try" in final_choice or "Escape" in final_choice:
            fail()
        elif "B" in final_choice or "Negotiate" in final_choice:
            fail()
        elif "C" in final_choice or "Fight" in final_choice:
            win()
        else:
            print("Try again")
            final_room()

#win function            
def win():
    print(f"""
          
__   _______ _   _   _    _ _____ _   _ _ 
\ \ / /  _  | | | | | |  | |_   _| \ | | |
 \ V /| | | | | | | | |  | | | | |  \| | |
  \ / | | | | | | | | |/\| | | | | . ` | |
  | | \ \_/ / |_| | \  /\  /_| |_| |\  |_|
  \_/  \___/ \___/   \/  \/ \___/\_| \_(_)
                                          
                                          
   The squirrels recognize your courage and have determined that you are a good nut!
   You have therefore, won this trophy {emoji.emojize(':trophy:',use_aliases=True)}!
      """)
    input("""
     <Press any key to exit>\n""")
    
#fail function    
def fail():
    print(f"""
     You have failed, the squirrels will throw you on the garbage chute {emoji.emojize(':open_mouth:',use_aliases=True)}.
     
     But thanks for playing {user_name} {emoji.emojize(':grimacing:',use_aliases=True)}!
    """)
    input("""
     <Press any key to exit>\n""")
    
    
    
main_door() #re-starts the game








```
